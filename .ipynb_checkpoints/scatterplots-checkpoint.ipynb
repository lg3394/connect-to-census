## 2-D visualizations (aka scatter plots)

It seems simple. It's not simple. mainly because...

**CORRELATION =/= CAUSATION**

Comparing variables to one another can be TRICKY!

Let's make some 2-D visualizations (aka scatter plots).

Be VERY careful how you phrase the headlines. Take time to think through what you want to say.




```python
%%R 
library(ggplot2)
# Calculate state averages
state_averages <- df %>%
  group_by(State) %>%
  summarize(
    avg_median_income = mean(median_income, na.rm = TRUE),
    avg_unhealthy_days = mean(UnhealthyDaysMonth, na.rm = TRUE),
    count = n()
  )

head(state_averages)

# Basic scatterplot first to check data
ggplot(state_averages, aes(x = avg_median_income, y = avg_unhealthy_days)) +
  geom_point() +
  labs(
    title = "Relationship Between Income and Physical Health by State",
    subtitle = "Average number of physically unhealthy days per month vs. median household income",
    x = "Average Median Income ($)",
    y = "Average Unhealthy Days per Month"
  ) +
  theme_minimal()
```

 ## 👉 Identifying outliers on the left and the right

```python
%%R
state_averages_clean <- state_averages %>%
  filter(!is.na(avg_median_income) & !is.na(avg_unhealthy_days))

# Calculate outlier thresholds using the 1.5 * IQR method
income_q1 <- quantile(state_averages_clean$avg_median_income, 0.25)
income_q3 <- quantile(state_averages_clean$avg_median_income, 0.75)
income_iqr <- income_q3 - income_q1
income_lower <- income_q1 - 1.5 * income_iqr
income_upper <- income_q3 + 1.5 * income_iqr

days_q1 <- quantile(state_averages_clean$avg_unhealthy_days, 0.25)
days_q3 <- quantile(state_averages_clean$avg_unhealthy_days, 0.75)
days_iqr <- days_q3 - days_q1
days_lower <- days_q1 - 1.5 * days_iqr
days_upper <- days_q3 + 1.5 * days_iqr

outliers <- state_averages_clean %>%
  filter(
    avg_median_income < income_lower | 
    avg_median_income > income_upper |
    avg_unhealthy_days < days_lower | 
    avg_unhealthy_days > days_upper
  )

print("Statistical outliers:")
print(outliers)
```

 ### This gave me the statistical outliers (only those on the right). I now also want to find the states on the left of the graph, especially the two with the highest number of unhealthy days per month

```python
%%R
worst_health_states <- state_averages %>%
  arrange(desc(avg_unhealthy_days)) %>%
  head(2)

# Display these states
print("States with highest average number of physically unhealthy days per month:")
print(worst_health_states)
```
